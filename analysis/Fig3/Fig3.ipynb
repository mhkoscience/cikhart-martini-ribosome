```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Open Sans'
lsp = 8
ls = 8
linew = 0.32
weight = 'normal'
#1-purple, 2-pink, 3-beige, 4-green
colors = ['#332288', '#CC6677', '#DDCC77', '#44AA79']
cm = 1/2.54
fig, axs = plt.subplots(nrows=6, figsize=(3.33, 3.1), sharex=True)


Rgyr_data = [
    [Rgyr_st1, Rgyr_st2, Rgyr_st3, Rgyr_st4],
    [Rgyr_so1, Rgyr_so2, Rgyr_so3, Rgyr_so4],
    [Rgyr_tw1, Rgyr_tw2, Rgyr_tw3, Rgyr_tw4],
    [Rgyr_lp1, Rgyr_lp2, Rgyr_lp3, Rgyr_lp4],
    [Rgyr_aa1, Rgyr_aa2, Rgyr_aa3, Rgyr_aa4],
    [Rgyr_sn1, Rgyr_sn2, Rgyr_sn3, Rgyr_sn4]
]

titles = ["STIFF", "SOFT", "SOFT2", "LIMP", "AA", "NONE"]

for idx, (ax, data, title) in enumerate(zip(axs, Rgyr_data, titles)):
    for j, dataset in enumerate(data):
        if idx == 4:
            ax.plot(
                dataset[:2000, 0] / 1000, dataset[:2000, 1] / 10,
                color=colors[j], linewidth=linew, alpha=1,label=f"{title} {chr(97 + j)}"
            )
        else:
            ax.plot(
                dataset[:, 0][::10] / 1000, dataset[:, 1][::10] / 10,
                color=colors[j], linewidth=linew, alpha=1,label=f"{title} {chr(97 + j)}"
            )
    
    
    ax.set_title(title, fontsize=ls, fontweight=weight)

    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylim(8.6,8.9)
    ax.set_yticks([8.6,8.8])
    ax.set_xticks([0,1000])
    ax.set_yticklabels([8.6,8.8], fontsize=8)
axs[2].set_ylabel("R$_\mathrm{gyr}$ [nm]", fontsize=lsp, fontweight=weight, labelpad=1)


axs[-1].set_xlabel('Time' + ' ' + '[$\mathrm{\mu}$s]', labelpad=-5, fontsize=8)
axs[-1].set_ylim(8.3,8.65)
axs[-1].set_yticks([8.3,8.6])
axs[-1].set_yticklabels([8.3,8.6])
axs[-1].set_xticklabels([0,1], fontsize=8)
handles, labels = axs[-1].get_legend_handles_labels()
plt.subplots_adjust(hspace=1.5)
plt.show()
```

```python
import numpy as np
plt.rcParams['font.family'] = 'Open Sans'

lw = 0.6
lsp = 8
ls = 8
l = 1.0
weight = 'normal'

colors = ['#332288', '#CC6677', '#DDCC77', '#44AA79']

fig, axs = plt.subplots(ncols=6, figsize=(3.33,0.5), sharex=False)

Rgyr_data = [
    [Rgyr_st1[:, 1]/10, Rgyr_st2[:, 1]/10, Rgyr_st3[:, 1]/10, Rgyr_st4[:, 1]/10],
    [Rgyr_so1[:, 1]/10, Rgyr_so2[:, 1]/10, Rgyr_so3[:, 1]/10, Rgyr_so4[:, 1]/10],
    [Rgyr_tw1[:, 1]/10, Rgyr_tw2[:, 1]/10, Rgyr_tw3[:, 1]/10, Rgyr_tw4[:, 1]/10],
    [Rgyr_lp1[:, 1]/10, Rgyr_lp2[:, 1]/10, Rgyr_lp3[:, 1]/10, Rgyr_lp4[:, 1]/10],
    [Rgyr_aa1[:2000, 1]/10, Rgyr_aa2[:2000, 1]/10, Rgyr_aa3[:2000, 1]/10, Rgyr_aa4[:2000, 1]/10],
    [Rgyr_sn1[:, 1]/10, Rgyr_sn2[:, 1]/10, Rgyr_sn3[:, 1]/10, Rgyr_sn4[:, 1]/10]
]

Rgyr_data_2 = [
    np.concatenate([Rgyr_st1[:, 1], Rgyr_st2[:, 1], Rgyr_st3[:, 1], Rgyr_st4[:, 1]]) / 10,
    np.concatenate([Rgyr_so1[:, 1], Rgyr_so2[:, 1], Rgyr_so3[:, 1], Rgyr_so4[:, 1]]) / 10,
    np.concatenate([Rgyr_tw1[:, 1], Rgyr_tw2[:, 1], Rgyr_tw3[:, 1], Rgyr_tw4[:, 1]]) / 10,
    np.concatenate([Rgyr_lp1[:, 1], Rgyr_lp2[:, 1], Rgyr_lp3[:, 1], Rgyr_lp4[:, 1]]) / 10,
    np.concatenate([Rgyr_aa1[:2000, 1], Rgyr_aa2[:2000, 1], Rgyr_aa3[:2000, 1], Rgyr_aa4[:2000, 1]]) / 10,
    np.concatenate([Rgyr_sn1[:, 1]/10, Rgyr_sn2[:, 1]/10, Rgyr_sn3[:, 1]/10, Rgyr_sn4[:, 1]/10])
]


titles = ["STIFF", "SOFT", "SOFT2", "LIMP", "AA", "NONE"]

for idx, (ax, data, title) in enumerate(zip(axs, Rgyr_data, titles)):
    hist, bins, patches = ax.hist(
        Rgyr_data_2[idx], bins='auto', orientation='vertical', linewidth=0,density=True,color=singlecolor, alpha=1
    )

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylim(0,30)
    ax.set_xlim(8.6,8.9)
    ax.set_title(titles[idx], fontsize=ls)
    ax.fill_between(bins[:-1],hist, lw=0.0, step="post")
    ax.set_yticks([10,20,30])
    ax.set_yticklabels([])
    ax.set_xticks([8.6,8.8])
    ax.set_xticklabels([8.6,8.8], rotation=0,fontsize=8)
    
axs[-1].set_xlim(8.3,8.66)
axs[-1].set_xticks([8.3,8.6])
axs[-1].set_xticklabels([8.3,8.6], fontsize=8)

axs[0].set_ylabel("pdf [arb. u.]", fontsize=lsp, fontweight=weight)
axs[2].set_xlabel("R$_\mathrm{gyr}$ [nm]", fontsize=lsp, fontweight=weight)
axs[0].set_yticks([0,10,20,30])
axs[0].set_yticklabels([0,10,20,30], fontsize=8)
plt.subplots_adjust(wspace=0.15)
plt.show()
```
