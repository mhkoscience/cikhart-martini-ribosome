```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
colors = ['#332288', '#CC6677', '#DDCC77', '#44AA79']
```

```python
singlecolor = '#4477AA'
```

```python
def load_angles_from_csv(path,classif,names=["stiff", "soft", "soft2", "limp"]):
    angles_loaded = {}
    for name in names:
        df = pd.read_csv(f"{path}angle_{classif}_{name}.csv")
        angles_loaded[name] = [df[col].values for col in df.columns]
    return angles_loaded

def load_angles_from_csv_aa(filename):
    df = pd.read_csv(filename) 
    angles_aa_loaded = {}
    for col in df.columns:
        idx = int(col.split("_")[1]) 
        angles_aa_loaded[idx] = df[col].values  
    return angles_aa_loaded
```

```python
angles_vectors_small_sci_point2 = load_angles_from_csv(path='./data/',classif='oneax')
angles_vectors_small_sci_point2_aa = load_angles_from_csv_aa("./data/angles_oneax_aa.csv")
```

```python
fig, axs = plt.subplots(
    ncols=6, nrows=1, figsize=(6, 0.5),
    sharex=True, gridspec_kw={'width_ratios': [1, 1, 1, 1,0.6,1], 'wspace': 0}
)
plt.rcParams['font.family'] = 'Open Sans'
axs[0].set_ylabel(r'$\mathrm{\Delta}\chi_{CG}$ [$^\circ$]', labelpad=0, fontsize=8)

axs[4].spines['top'].set_visible(False)
axs[4].spines['right'].set_visible(False)
axs[4].spines['left'].set_visible(False)
axs[4].spines['bottom'].set_visible(False)
axs[4].set_xticks([])
axs[4].set_yticks([])

for ix in range(4):
    axs[ix].set_xticks([0,1])
    axs[ix].set_xticks([0,1])
    
    axs[ix].set_xticklabels([0,1], size=8)
    axs[ix].set_ylim(-10,10)
    
    axs[ix].set_yticks([-10,10])
    axs[ix].set_yticklabels([-10,10], size=8)
    axs[ix].set_xlabel('Time' + ' ' + '[$\mathrm{\mu}$s]', labelpad=0, fontsize=8)
    axs[ix].spines['top'].set_visible(False)
    axs[ix].spines['right'].set_visible(False)
    if ix > 0:
        axs[ix].set_yticklabels([])
axs[0].set_yticks([-10,0,10])
axs[0].set_yticklabels([-10,0,10], size=8)
axs[5].set_yticks([-10, 0,10])
axs[5].set_xticks([0, 1])
axs[5].set_xticklabels([0,1], size=8)
axs[5].set_yticklabels([-10,0,10], size=8)
axs[5].set_ylim(-10, 10)
axs[5].set_ylabel(r'$\mathrm{\Delta}\chi_{AA}$ [$^\circ$]', labelpad=0,     fontsize=8)
axs[5].set_xlabel('Time' + ' ' + '[$\mathrm{\mu}$s]', labelpad=0, fontsize=8)
axs[5].spines['top'].set_visible(False)
axs[5].spines['right'].set_visible(False)
axs[4].tick_params(axis='x', bottom=False, labelbottom=False)

lw = 0.32
for i in range(4):
    axs[0].plot(np.array(range(2000))[::20]/2000,
                angles_vectors_small_sci_point2['stiff'][i][::20]-angles_vectors_small_sci_point2['stiff'][i][0], colors[i], linewidth=lw)
    axs[0].set_title('STIFF', fontsize=8)

    axs[1].plot(np.array(range(2000))[::20]/2000,
                angles_vectors_small_sci_point2['soft'][i][::20]-angles_vectors_small_sci_point2['soft'][i][0], colors[i], linewidth=lw)
    axs[1].set_title('SOFT', fontsize=8)

    axs[2].plot(np.array(range(2000))[::20]/2000,
                angles_vectors_small_sci_point2['soft2'][i][::20]-angles_vectors_small_sci_point2['soft2'][i][0], colors[i], linewidth=lw)
    axs[2].set_title('SOFT2', fontsize=8)

    axs[3].plot(np.array(range(2000))[::20]/2000,
                angles_vectors_small_sci_point2['limp'][i][::20]-angles_vectors_small_sci_point2['limp'][i][0], colors[i], linewidth=lw)
    axs[3].set_title('LIMP', fontsize=8)

    axs[5].plot(np.array(range(2000))[::20]/2000, angles_vectors_small_sci_point2_aa[i][::20]-angles_vectors_small_sci_point2_aa[i][0], colors[i], linewidth=lw)
    axs[5].set_title('AA', fontsize=8)

plt.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0.00, hspace=0)
```
